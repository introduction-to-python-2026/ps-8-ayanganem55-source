```python
import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv("freelance_jobs_2025.csv")
print(df.head)

# Convert 'Annual openings (BLS)' to numeric, removing commas
df['Annual openings (BLS)'] = df['Annual openings (BLS)'].str.replace(',', '').astype(int)

df_cleaned = df[df[ "Annual openings (BLS)"] >= 10]


x = df["Job title"]
y = df["Annual openings (BLS)"]
plt.figure(figsize=(20,5))

plt.plot(x,y, alpha=0.6)
plt.xlabel("job title")
plt.ylabel("Annual openings (BLS)")
plt.show()
```
